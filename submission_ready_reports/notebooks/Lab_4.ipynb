# Simulation of Drunk Man Random Walk

## Objective
Simulate the experiment and interpret the outcome.


## Simulation and Modelling Framework
**System/Process Model:** Discrete-time stochastic random walk model on a 2D lattice.

**Key Assumptions:** Each step is independent with equal probability among four directions.

**Simulation Method:** Single-path trajectory simulation plus multi-trial ensemble statistics.

**Validation Strategy:** Verify near-zero mean displacement trend and sqrt(n)-type distance growth behavior.


## Theory
A drunk man random walk is a stochastic process where each step direction is random.
In a 2D walk, each step can move up, down, left, or right with equal probability.
Although the expected displacement is near zero, expected distance from origin grows with sqrt(n).

## Mathematical Equations
$$\vec{r}_n = \sum_{i=1}^{n} \vec{s}_i$$

$$E[\vec{r}_n] \approx 0, \qquad E[|\vec{r}_n|] \propto \sqrt{n}$$

## Required Topics
1. **Random walk model**
2. **Stochastic process**
3. **Mean distance from origin**
4. **Simulation statistics**

## Brief Explanation of Required Topics
**1. Random walk model**
This is the mathematical model of the drunk man's step-by-step movement.

**2. Stochastic process**
Each step is random, so the experiment is governed by probability rather than fixed paths.

**3. Mean distance from origin**
A key performance metric showing how far, on average, the walker ends up after many steps.

**4. Simulation statistics**
Multiple trials and summary statistics are needed to draw reliable conclusions from randomness.

## How to Run
Execute the code cell to generate results and graph.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(7)
n_steps = 1000
n_trials = 500

# Single trajectory
directions = np.random.randint(0, 4, n_steps)
step_map = np.array([[1,0],[-1,0],[0,1],[0,-1]])
steps = step_map[directions]
path = np.vstack(([0,0], np.cumsum(steps, axis=0)))

# Ensemble statistics
final_distances = []
for _ in range(n_trials):
    d = np.random.randint(0, 4, n_steps)
    s = step_map[d]
    final_pos = np.sum(s, axis=0)
    final_distances.append(np.linalg.norm(final_pos))

print(f"Mean final distance from origin: {np.mean(final_distances):.4f}")
print(f"Std. dev. of final distance: {np.std(final_distances):.4f}")

plt.figure(figsize=(6,6))
plt.plot(path[:,0], path[:,1], color='teal')
plt.scatter([0], [0], c='red', label='Start')
plt.title('2D Drunk Man Random Walk (Single Trial)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```

## Expected Result and Conclusion

With 1000 steps and 500 trials, mean final distance from origin is 27.7225 and standard deviation is 14.1306.

**Conclusion:** The random walk confirms stochastic behavior and the growth of average distance with step count.
