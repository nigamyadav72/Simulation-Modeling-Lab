# Simulation of Discharging of a Capacitor

## Objective
Simulate the experiment and interpret the outcome.


## Simulation and Modelling Framework
**System/Process Model:** Deterministic first-order RC discharge model with exponential decay.

**Key Assumptions:** Ideal resistor and capacitor, no source during discharge, constant R and C values.

**Simulation Method:** Time-domain simulation by evaluating analytical decay equation over sampled times.

**Validation Strategy:** Check simulated voltage at t = τ against theoretical value 0.368V0.


## Theory
In an RC circuit, capacitor voltage decreases exponentially during discharge.
The time constant is represented by the symbol τ and controls how quickly the voltage drops.
At time t = τ, the voltage becomes about 36.8% of the initial value.
This simulation verifies the analytical discharge model through numerical plotting.

## Mathematical Equations
$$V(t) = V_0 e^{-t/(RC)}$$

$$\tau = RC$$

$$V(\tau) = V_0 e^{-1} \approx 0.368V_0$$

## Brief Explanation of Required Topics
**1. RC circuits**
The capacitor discharge model directly comes from RC circuit behavior.

**2. Exponential decay**
Voltage reduction over time follows an exponential curve, which the simulation must reproduce.

**3. Time constant concept**
The value τ = RC gives a physical timescale for how fast discharge occurs.

**4. Scientific plotting**
A clear voltage-time plot is needed to verify model behavior and present results in reports.

## How to Run
Execute the code cell to generate results and graph.


```python
import numpy as np
import matplotlib.pyplot as plt

R = 1000.0      # ohm
C = 100e-6      # farad
V0 = 10.0       # volt

tau = R * C
t = np.linspace(0, 1.0, 500)
V = V0 * np.exp(-t / tau)

print(f"Time constant τ = RC: {tau:.4f} s")
print(f"Voltage at t=τ: {V0*np.exp(-1):.4f} V")

plt.figure(figsize=(8,4))
plt.plot(t, V, color='darkorange')
plt.axvline(tau, linestyle='--', color='gray', label='τ = RC')
plt.title('Capacitor Discharging Curve')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.grid(True)
plt.legend()
plt.show()
```

## Expected Result and Conclusion

For R = 1000 ohm, C = 0.000100 F and V0 = 10.00 V, τ = 0.1000 s. Voltage at t=τ is 3.6788 V.

**Conclusion:** The voltage decays exponentially and matches the theoretical RC discharge equation.
