# Approximation of Pi Using Monte Carlo Simulation

## Objective
Simulate the experiment and interpret the outcome.


## Simulation and Modelling Framework
**System/Process Model:** Stochastic geometric probability model using random sampling in a unit square.

**Key Assumptions:** Independent and uniformly distributed random points; large sample approximates true area ratio.

**Simulation Method:** Monte Carlo simulation with indicator counting for quarter-circle membership.

**Validation Strategy:** Measure absolute error between estimated pi and reference pi; observe convergence with high N.


## Theory
Monte Carlo simulation estimates unknown values by repeated random sampling.
To approximate pi, random points are generated inside a unit square.
The ratio of points inside a quarter-circle to the total points estimates the area ratio.
Multiplying that ratio by 4 gives an approximation of pi.

## Mathematical Equations
$$x_i, y_i \sim U(0,1)$$

$$I_i = \begin{cases}1, & x_i^2 + y_i^2 \le 1 \\ 0, & \text{otherwise}\end{cases}$$

$$\hat{\pi} = 4\left(\frac{1}{N}\sum_{i=1}^{N} I_i\right)$$

## Required Topics
1. **Random number generation**
2. **Geometric probability**
3. **Monte Carlo method**
4. **Error analysis**

## Brief Explanation of Required Topics
**1. Random number generation**
Monte Carlo simulation depends on random samples, so quality random numbers are essential.

**2. Geometric probability**
Pi is estimated from the geometry of points inside and outside a quarter circle.

**3. Monte Carlo method**
This is the simulation framework where repeated random trials approximate a mathematical constant.

**4. Error analysis**
Comparing estimate with true pi quantifies simulation accuracy and convergence quality.

## How to Run
Execute the code cell to generate results and graph.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
N = 100000
x = np.random.rand(N)
y = np.random.rand(N)

inside = (x**2 + y**2) <= 1
pi_est = 4 * np.mean(inside)
abs_error = abs(np.pi - pi_est)

print(f"Estimated pi: {pi_est:.6f}")
print(f"Absolute error: {abs_error:.6f}")

sample = 3000
plt.figure(figsize=(6,6))
plt.scatter(x[:sample][inside[:sample]], y[:sample][inside[:sample]], s=3, c='green', label='Inside')
plt.scatter(x[:sample][~inside[:sample]], y[:sample][~inside[:sample]], s=3, c='crimson', label='Outside')
plt.title('Monte Carlo Estimation of Pi (Sample View)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```

## Expected Result and Conclusion

For N = 100000 random points, the estimated value of pi is 3.137600. The absolute error is 0.003993.

**Conclusion:** The Monte Carlo estimate converges near pi, and increasing sample size improves accuracy.
