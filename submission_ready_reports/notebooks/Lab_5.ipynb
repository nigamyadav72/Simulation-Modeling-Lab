# Simulation of Chi-Squared Test

## Objective
Simulate the experiment and interpret the outcome.


## Simulation and Modelling Framework
**System/Process Model:** Statistical simulation/analysis model for categorical frequency comparison.

**Key Assumptions:** Independent observations and adequate expected counts in each category.

**Simulation Method:** Compute chi-squared statistic and p-value from observed and expected frequencies.

**Validation Strategy:** Decision consistency based on significance level alpha and interpreted p-value.


## Theory
The chi-squared goodness-of-fit test compares observed category frequencies with expected frequencies.
The test statistic measures the overall deviation between observed and expected counts.
A p-value larger than the significance level indicates that observed data is consistent with the expected model.

## Mathematical Equations
$$\chi^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$$

$$H_0: O_i \text{ follows expected distribution}, \quad H_1: O_i \text{ does not}$$

## Required Topics
1. **Goodness-of-fit testing**
2. **Null and alternative hypotheses**
3. **Chi-squared statistic**
4. **p-value based decision**

## Brief Explanation of Required Topics
**1. Goodness-of-fit testing**
This checks whether observed simulation/experimental categories match an expected model.

**2. Null and alternative hypotheses**
These define what it means for data to fit or not fit the expected distribution.

**3. Chi-squared statistic**
It quantifies how much observed counts deviate from expected counts.

**4. p-value based decision**
The p-value provides the decision rule used to accept or reject model consistency.

Execute the code cell to generate results and graph.


```python
import matplotlib.pyplot as plt

# Data for plotting
faces = ['1', '2', '3', '4', '5', '6']

# Create bar positions
x = np.arange(len(faces))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, observed, width, label='Observed', color='skyblue')
bars2 = ax.bar(x + width/2, expected, width, label='Expected', color='lightcoral')

# Add labels and title
ax.set_xlabel('Die Face')
ax.set_ylabel('Frequency')
ax.set_title('Chi-Square Goodness-of-Fit Test: Observed vs Expected Frequencies')
ax.set_xticks(x)
ax.set_xticklabels(faces)
ax.legend()

# Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}', ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([16, 19, 17, 15, 21, 18])   # counts for die faces 1..6
total = observed.sum()
expected = np.full(6, total / 6)

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)
alpha = 0.05

print(f"Observed counts: {observed.tolist()}")
print(f"Expected counts: {expected.tolist()}")
print(f"Chi-squared statistic: {chi2_stat:.4f}")
print(f"p-value: {p_value:.4f}")

if p_value < alpha:
    print("Decision: Reject H0 (distribution differs significantly).")
else:
    print("Decision: Fail to reject H0 (distribution is consistent with expectation).")
```

## Expected Result and Conclusion

Observed counts = [15, 18, 16, 14, 20, 17] with total 100. Chi-squared statistic = 1.4000, p-value = 0.9243. Decision at alpha=0.05: Fail to reject H0 (distribution is consistent with expected).

**Conclusion:** The chi-squared framework effectively evaluates whether observed categorical data fits the expected model.
